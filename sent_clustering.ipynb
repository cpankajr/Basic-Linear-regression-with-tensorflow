```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.feature_extraction import text 
import xlrd 
import csv

def strip_non_ascii(string):
    ''' Returns the string without non ASCII characters'''
    stripped = (c for c in string if 0 < ord(c) < 127)
    return ''.join(stripped)

documents = []

loc = ("D:/Downloads/chat-history.xlsx") 
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(0)
```

```python
import re
documents = []

for i in range(1, 3940):
    text_msg = strip_non_ascii(sheet.cell_value(i, 10))
    message = str(text_msg)
    message=message.lower()
    message = re.sub(r'[^\w\s]','',message)
#     print(message)
    documents.append(message)

documents=list(set(documents))    
my_additional_stop_word_list = ["hello", "hi", "yes", 'u','m']
stop_words = text.ENGLISH_STOP_WORDS.union(my_additional_stop_word_list)

vectorizer = TfidfVectorizer(stop_words=set(stop_words))
X = vectorizer.fit_transform(documents)

true_k = 10
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

output_file_content = []

for i in documents:
    if(i in stop_words):
        continue
    Y = vectorizer.transform([i])
    prediction = model.predict(Y)
#     print(i, Y,prediction)
    temp = [i, prediction[0]]
    output_file_content.append(temp)

with open('cluster_10.csv', 'w') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerows(output_file_content)

csvFile.close()
```

```python
loc = ("D:/Downloads/chat-history.xlsx") 
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(0)
documents = []
for i in range(1, 3940):
    text_msg = strip_non_ascii(sheet.cell_value(i, 10))
    message = str(text_msg)
    msg=message.lower()
    documents.append(msg)
my_additional_stop_word_list = ["hello","yes", "hi", "hi..."]
stop_words = text.ENGLISH_STOP_WORDS.union(my_additional_stop_word_list)
```

```python
part_of_speech=['NN','NNS', 'NNP', 'NNPS', 'JJ', 'JJR', 'JJS']
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize 
```

```python
keywrd=[]
bigramfdist = nltk.FreqDist()
trigramfdist = nltk.FreqDist()
for i in documents:
    wordsList = nltk.word_tokenize(i)  
    wordsList = [w for w in wordsList if not w in stop_words]
    bgs = nltk.ngrams(wordsList, 2)
    
    bigramfdist.update(bgs)

    tgs = nltk.ngrams(wordsList, 3)
    trigramfdist.update(tgs)
    
    tagged = nltk.pos_tag(wordsList)
    if (tagged !=[]):
        for j in range(len(tagged)):
            if(tagged[j][1] in part_of_speech):
                keywrd.append(tagged[j][0])
# print(keywrd)
```

```python
for word, frequency in bigramfdist.most_common(50):
    
    print(u'{};{}'.format(word, frequency))
```

```python
for word, frequency in trigramfdist.most_common(50):
    print(u'{};{}'.format(word, frequency))
```

```python
fdist = nltk.FreqDist(keywrd)
```

```python
for word, frequency in fdist.most_common():
    print(u'{};{}'.format(word, frequency))
```

```python
nouns=list(set(keywrd))
```

```python
bg_list=[]
for word, frequency in bigramfdist.most_common():
    temp=list(word)
    if set(temp)<set(nouns):
        bg_list.append([word,frequency])
```

```python
file_content=[]

for word, frequency in  trigramfdist.most_common():
    print(u'{};{}'.format(word, frequency))
    temp = ["".join(word), frequency]
    file_content.append(temp)
with open('frequent_trigram.csv', 'w') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerows(file_content)
```
